```python
# basic imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# make plots look nice
sns.set(style='whitegrid')
```

```python
df = pd.read_csv('/content/imdb_movies.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

*Clean & Prepare Data*

```python
# keep only relevant columns (if names differ, adjust)
df = df[['names','genre','date_x','score']]

# drop missing years or ratings
df = df.dropna(subset=['date_x','score'])

# Convert 'date_x' to datetime and extract the year
df['Released_Year'] = pd.to_datetime(df['date_x'], errors='coerce').dt.year

# ensure numeric types
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df = df[df['Released_Year'] >= 1950]   # optional filter
```

```python
df['genre'] = df['genre'].fillna('Unknown')
df['genre'] = df['genre'].str.split(', ')
df = df.explode('genre')
```

# Analyze Ratings by Year

```python
ratings_by_year = df.groupby('Released_Year')['score'].mean().reset_index()
plt.figure(figsize=(12,5))
sns.lineplot(x='Released_Year', y='score', data=ratings_by_year)
plt.title('Average IMDb Rating by Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

# Analyze Ratings by Genre

```python
ratings_by_genre = df.groupby('genre')['score'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
ratings_by_genre.plot(kind='bar', color='skyblue')
plt.title('Average IMDb Rating by Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

# Top Movies

```python
top10 = df.sort_values('score', ascending=False).head(10)
top10[['names','Released_Year','genre','score']]
```

# Distribution of Ratings

```python
plt.figure(figsize=(8,4))
sns.histplot(df['score'], bins=20, kde=True)
plt.title('Distribution of IMDb Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.show()
```
